### 웹 크롤링
- 웹 크롤러는 웹문서, 이미지 등을 주기적으로 쉽하여 자동으로 데이터베이스화하는 프로그램
- 크롤러가 하는 직업을 웹 크롤링
- requests와 beautifulsoup4 패키지 사용

### 1. 웹문서 전체 가져오기
- urllib.request

```python
from urllib.request import urlopen
import requests
from bs4 import BeautifulSoup as bs
```

```python
html = urlopen('https://www.naver.com/')
soup = bs(html, 'html.parser')
print(soup)
```

- requests

```python
html = requests.get('https://www.naver.com/')
soup = bs(html.text, 'html.parser')
print(soup)
```

### 2. html 태그에서 원하는 정보 추출하기
- html 태그를 이용하는 방법
- CSS Selector를 이용하는 방법

- html 태그: 
    - find('태그') - 첫번째 태그만 검색 / 
    - find_all('태그') - 태그 전체를 검색(list로 반환)

```python
## class를 이용한 태그 검색

find_div = soup.find('div', class_ = 'group_nav')
find_div

find_div = soup.find('div', id='NM_FAVORITE' )
find_div
```

```python
# find_all 사용
find_div = soup.find('div', class_ = 'group_nav')
find_lst = find_div.find_all('li')
for item in find_lst:
    print(item.get_text()) # 모든 태그를 제거하고 텍스트만 남김
    print(item.find('a')['href']) # 태그의 속성값을 추출

##item.text => 태그 내에 있는 텍스트를 추출
```

### [크롤링 연습]
- 네이버 지식인에서 '선릉역'으로 검색 후 검색 결과 가져오기

```python
url = 'https://kin.naver.com/search/list.naver?query=%EC%84%A0%EB%A6%89%EC%97%AD'

html = urlopen(url)
soup = bs(html, 'html.parser')

#css를 이용한 태그 검색
ul_soup = soup.select('#s_content > div.section > ul')

#검색된 태그에서 다시 li 전체를 검색
li_soup = ul_soup[0].find_all('li') 

# 검색된 li에서 원하는 텍스트 크롤링(추출)
print(li_soup[0].find('dt').text)  ##제목
print(li_soup[0].find('dd').get_text())  ##날짜
print(li_soup[0].select('#s_content > div.section > ul > li:nth-child(1) > dl > dd:nth-child(3)')[0].text) #요약내용
print(li_soup[0].find_all('dd')[1].get_text())
```

### [미션] 1페이지 검색 결과에서 전체 내용 가져오기
- 제목, 등록일, 요약 부분을 각각 가져와 DataFrame으로 저장

```python
import pandas as pd

url = 'https://kin.naver.com/search/list.naver?query=%EC%84%A0%EB%A6%89%EC%97%AD'
html = urlopen(url)
soup = bs(html, 'html.parser')
ul_soup = soup.select('#s_content > div.section > ul')
li_soup = ul_soup[0].find_all('li') 

nav_kin = []

for i in li_soup:
    title = i.find('dt').get_text().replace('\n','')
    date = i.find_all('dd')[0].get_text()
    doct = i.find_all('dd')[1].get_text()
    
    nav_kin.append({'제목':title,'등록일':date,'요약':doct})

# print(nav_kin)
df = pd.DataFrame(nav_kin)
df
    
```

### 3.검색어를 이용한 크롤링

### 4.검색된 내용 전체(페이지 이동)

```python
# 검색어 입력받기
text= input('검색어 입력: ')
nav_kin = []
for page in range(1,11):
    url = 'https://kin.naver.com/search/list.naver?query=' + parse.quote(text) + '&page=' + str(page)

    html = urlopen(url)
    soup = bs(html, 'html.parser')
    ul_soup = soup.select('#s_content > div.section > ul')
    li_soup = ul_soup[0].find_all('li') 

    

    for i in li_soup:
        title = i.find('dt').get_text().replace('\n','')
        date = i.find_all('dd')[0].get_text()
        doct = i.find_all('dd')[1].get_text()

        nav_kin.append({'제목':title,'등록일':date,'요약':doct})

# print(nav_kin)
df = pd.DataFrame(nav_kin)
df
```
